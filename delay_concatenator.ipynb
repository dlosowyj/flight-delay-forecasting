```python
import pandas as pd
import os
import fnmatch
from collections import defaultdict
```

```python
# group files by airport for easier sorting
def group_files(directory_path):
    grouped_files = defaultdict(list)
    prefix_length = 3 # airport codes are 3 letters long

    # Iterate over all entries in the specified directory
    try:
        for filename in os.listdir(directory_path):
            # Ensure the entry is actually a file (and not a subdirectory)
            full_path = os.path.join(directory_path, filename)
            if os.path.isfile(full_path):
                # Get the first letters of the filename (in this case 3 for the airport codes)
                prefix = filename[:prefix_length]
                # Add the filename to the list corresponding to its prefix
                grouped_files[prefix].append(full_path)
    except FileNotFoundError:
        print(f"Error: Directory not found at '{directory_path}'")
        return {}
    except Exception as e:
        print(f"An error occurred: {e}")
        return {}

    return grouped_files

def concat_files(grouped_files):
    for key, val_list in grouped_files.items():
        df = pd.DataFrame()
        for val in val_list:
            df = pd.concat([df, pd.read_csv(val, skiprows=7)])

        df.to_csv('data\\delays\\concatenated_delays\\' + str(key) + '_delays.csv')
```

```python
grouped_files = group_files('data\\delays')
```

```python
concat_files(grouped_files)
```
